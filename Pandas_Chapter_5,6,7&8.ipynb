## Chapter_8 Crosstab PivotTable Groupby

```python
insurance=pd.read_csv("insurance.csv")
```

```python
insurance
```

```python
pd.crosstab(index=insurance["region"],columns=insurance["smoker"])
```

```python
pd.pivot_table(insurance,
    values="charges",
    index="region",
    columns="smoker",aggfunc="sum")             ## Groupby func and Merge func assignment
```
